```python
# !pip install undetected-chromedriver

# Selenium : 수집용, 매크로!!!!!!!!!!!
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
driver = uc.Chrome() # 크롬창을 하나 열라는 명령
time.sleep(3) # 3초정도 쉬었다가
driver.get('https://www.naver.com/') # 네이버로 들어가라는 명령
```

```python
search = driver.find_element(By.CLASS_NAME, 'search_input')
# search # 뭔가 나오면 찾았다는 것.
search.send_keys('대구은행') # 대구은행을 입력하고
time.sleep(3) # 3초정도 쉬었다가
search.send_keys(Keys.RETURN) # 엔터를 누르도록
```

```python
# 인터파크 전시
driver = uc.Chrome() # 크롬창을 하나 열라는 명령
driver.get('https://tickets.interpark.com/goods/24015987') 
time.sleep(3) # 3초정도 쉬었다가
driver.find_element(By.CLASS_NAME, 'popupCloseBtn.is-bottomBtn').click() # 공백은 .으로 이어주기 // click()으로 닫기 버튼 누르기.
time.sleep(1)

A = driver.find_element(By.CSS_SELECTOR,'ul.navList').find_elements(By.CSS_SELECTOR, 'li.navItem') # 태그명이랑 클래스 이름을 같이 적을 수 있음
# len(A)
for i in A:
    if '이용후기' in i.text : #이용후기가 있으면 클릭하도록
        i.click()
```

```python
import pandas as pd
box = []
for i in driver.find_elements(By.CSS_SELECTOR, 'li.bbsItem') :
    title = i.find_element(By.CSS_SELECTOR,'strong.bbsTitleText').text.strip()
    content = i.find_element(By.CSS_SELECTOR, 'p.bbsText').text.strip()
    star = i.find_element(By.CSS_SELECTOR, 'div.prdStarIcon').get_attribute('data-star') # 속성에 접근기기
    name = i.find_element(By.CSS_SELECTOR, 'span.name').text.strip()
    date = i.find_elements(By.CSS_SELECTOR, 'li.bbsItemInfoList')[1].text.strip() # 리스트로 다 가져와서 1번째 인덱스에 있는 날짜 가져기기
    box.append([title,content,star,name,date])

box
```

```python
pd.DataFrame(box, columns=['제목','내용','작성자','별점','날짜'])
```

```python

box = []
while True :
    numbers = driver.find_element(By.CSS_SELECTOR, 'ol.pageNumWrap').find_elements(By.CSS_SELECTOR, 'li')
    for k in numbers :
        k.click()
        time.sleep(2)

        for i in driver.find_elements(By.CSS_SELECTOR, 'li.bbsItem') :
            title = i.find_element(By.CSS_SELECTOR,'strong.bbsTitleText').text.strip()
            content = i.find_element(By.CSS_SELECTOR, 'p.bbsText').text.strip()
            star = i.find_element(By.CSS_SELECTOR, 'div.prdStarIcon').get_attribute('data-star') # 속성에 접근기기
            name = i.find_element(By.CSS_SELECTOR, 'span.name').text.strip()
            date = i.find_elements(By.CSS_SELECTOR, 'li.bbsItemInfoList')[1].text.strip() # 리스트로 다 가져와서 1번째 인덱스에 있는 날짜 가져기기
            box.append([title,content,star,name,date])

    try :
        driver.find_element(By.CSS_SELECTOR, 'a.pageNextBtn.pageArrow').click()
        time.sleep(2)
    except :
        break
```

```python
result = pd.DataFrame(box, columns=['제목','내용','작성자','별점','날짜'])
result.to_excel('인터파크.xlsx',index = False)
```
